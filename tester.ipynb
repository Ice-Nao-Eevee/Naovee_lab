```python
# 1. Tentukan data poin masing-masing
poin_naovee = 230400
poin_evoly = 200320
poin_per_match = 300

# 2. Hitung total poin gabungan
total_poin = poin_naovee + poin_evoly

# 3. Hitung jumlah pertandingan
# Kita gunakan pembagian bulat (//) jika asumsinya tidak ada setengah pertandingan
total_pertandingan = total_poin / poin_per_match

# 4. Tampilkan hasil
print(f"Total poin Naovee dan Evoly: {total_poin}")
print(f"Total pertandingan yang telah dimainkan: {total_pertandingan:.2f} pertandingan")
```

```python
# Data awal
poin_naovee = 230400
poin_evoly = 200320
poin_per_match = 300

# Hitung pertandingan masing-masing
match_naovee = poin_naovee / poin_per_match
match_evoly = poin_evoly / poin_per_match

# Tampilkan hasil dengan pembulatan 2 angka di belakang koma agar rapi
print(f"Naovee bermain sebanyak: {match_naovee:.2f} pertandingan")
print(f"Evoly bermain sebanyak: {match_evoly:.2f} pertandingan")

# Total gabungan untuk verifikasi
print("-" * 30)
print(f"Total gabungan: {match_naovee + match_evoly:.2f} pertandingan")
```
